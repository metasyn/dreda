```python
from sklearn import datasets
from sklearn.cluster import KMeans

iris = datasets.load_iris()
X = iris.data
Y = iris.target

estimator = KMeans(3)
estimator.fit(X)

# Create Data Frame of x, y, z (petal length, sepal length, pedal width or something)
# Transpose to make the rows the points (thats the T)
# Multiply by 100 to make it a nice size
iris_df = pd.DataFrame([X[:, 3], X[:, 0], X[:, 2]]).T
```

```python
iris_df.head()
```

```python
# Add colors
iris_df['c'] = estimator.labels_
```

```python
iris_df.c.value_counts()
```

```python
iris_df.to_json('iris.json')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.palplot(sns.color_palette('husl', 30))
```

```python
sns.color_palette('husl', 30)
```

```python
colors = sns.color_palette('husl', 30)
```

```python
colors
```

```python
rgbs = []
for (r,g,b) in colors:
    r *= 255
    g *= 255
    b *= 255
    rgbs.append((int(r),int(g),int(b)))
```

```python
rgbs
```

```python
hexes = ["#{0:02x}{1:02x}{2:02x}".format(r,b,g) for (r,g,b) in rgbs]
```

```python
dataprojector_data = pd.read_json("./data.json")
```

```python
dataprojector_data.head()
```

```python
dataprojector_data['y'] = dataprojector_data.points.apply(lambda x: float(str(x).split(',')[0].split()[1][1:-1]))
dataprojector_data['x'] = dataprojector_data.points.apply(lambda x: float(str(x).split(',')[1].split()[1][1:-1]))
dataprojector_data['z'] = dataprojector_data.points.apply(lambda x: float(str(x).split(',')[2].split()[1][1:-1]))
dataprojector_data['cid'] = dataprojector_data.points.apply(lambda x: float(str(x).split(',')[3].split()[1].replace('\'', '').replace('}', '')))
```

```python
dataprojector_data
```

```python
dp_data_clean = dataprojector_data.drop('points', 1)
```

```python
dp_data_clean.head()
```
